```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-100, 100, 101)
trian = 100 - np.abs(t)
x1 = np.zeros(1001)
x2 = np.zeros(1001)
x1[450:551] = trian 
x2[440:541] = trian 
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x1, label='x1')
plt.plot(x2, label='x2')
plt.legend()
plt.grid()
plt.show()
```

```python
# forward transform:
f_x1 = np.fft.fft(x1)
f_x2 = np.fft.fft(x2)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(np.fft.fftshift(np.real(f_x1))[450:551], label='f_x1')
plt.plot(np.fft.fftshift(np.real(f_x2))[450:551], label='f_x2')
plt.legend()
plt.grid()
plt.show()
```

```python
# multiplication and inv transform:
x1_conv_x2 = np.fft.ifft(f_x1 * np.conj(f_x2) / np.abs(f_x1 * np.conj(f_x2)))
```

```python
plt.figure(figsize=(10, 5))
plt.plot(np.abs(x1_conv_x2))
plt.grid()
plt.title(f'x1 conv x2: argmax = {np.argmax(np.abs(x1_conv_x2))}')
plt.show()
```
